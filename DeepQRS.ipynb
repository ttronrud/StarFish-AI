```python
import wfdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.utils import to_categorical
from keras.layers import Conv1D
from keras.layers import MaxPooling1D
import random
import math

##MAIN CONSTANTS TO ADJUST
num_sec = 6 #length of time to send through model
fs = 360 #sample rate
min_RR = 0.3 #minimum length of time between Rs in seconds
#0.3 ~200bpm
####


abnormal = ['L','R','V','/','A','f','F','j','a','E','J','e','S']

data_path = 'MIT-BIH_Arrythm/'

# list of patients
pts = ['100','101','103','105','106',
       '108','109','111','112','113','114','115','116',
       '117','118','119','121','122','123','124','200',
       '201','202','203','205','207','208','209','210',
       '212','213','214','215','219','220','221',
       '222','223','228','230','231','232','233','234']

pt_files = [data_path + pt for pt in pts]
df = pd.DataFrame()

for pt in pts:
    file = data_path + pt
    annotation = wfdb.rdann(file, 'atr')
    sym = annotation.symbol
    
    values, counts = np.unique(sym, return_counts=True)
    df_sub = pd.DataFrame({'sym':values, 'val':counts, 'pt':[pt]*len(counts)})
    df = pd.concat([df, df_sub],axis = 0)

def wfdb_load_ecg(file):
    record = wfdb.rdrecord(file)
    annotation = wfdb.rdann(file, 'atr')
    
    # extract the signal
    p_signal = record.p_signal
    
    # verify frequency is 360
    assert record.fs == 360, 'sample freq is not 360'
    
    # extract symbols and annotation index
    atr_sym = annotation.symbol
    atr_sample = annotation.sample
    
    return p_signal, atr_sym, atr_sample 

def gen_dataset(pt_list, sample_sec = 3, fs=360., targets=abnormal + ['N'], drop_to = -1):
    N = int(sample_sec*fs)
    
    Xs = [] #signal samples
    Ys = [] #number of N annotations contained within the signal
    
    for pt in pt_list:
        sig,symbols,sym_sample = wfdb_load_ecg(pt)
        sig = sig[:,0]
        nuDim = int(math.floor(sig.shape[0]/N)) #how many recordings of length N can we get?
        numN = np.zeros(nuDim)
        #split the signal into equal chunks using reshape
        SplitSig = sig[:N*nuDim].reshape((nuDim,N))
        #now, get number of Ns within signal
        for inde, offset in enumerate(range(nuDim)):
            sym_fil = (sym_sample > offset*N)&(sym_sample < (offset+1)*N)
            _n = 0
            keeps = len(np.array(symbols)[sym_fil])
            if drop_to < 0: #can drop up to all
                keeps = len(np.array(symbols)[sym_fil])-np.random.randint(len(np.array(symbols)[sym_fil])-1)
            elif drop_to > 0:
                keeps = len(np.array(symbols)[sym_fil])-np.random.randint(drop_to)
            for i,s in enumerate(np.array(symbols)[sym_fil]):
                if i < keeps:
                    if np.isin(s,targets):
                        _n += 1
                else:
                    samnum = sym_sample[sym_fil][i]
                    SplitSig[inde][samnum-offset*N-75:] = 0
                    break
            #numN[inde] = _n
            ssig_baseline = np.convolve(SplitSig[inde],np.ones(64)/64., mode = 'same')
            debaselined_ssig = SplitSig[inde,:ssig_baseline.shape[0]] - ssig_baseline
            Xs.append(debaselined_ssig)
            Ys.append([_n])
    return np.array(Xs), np.array(Ys)
    
ssig,nN = gen_dataset(pt_files[:10],sample_sec=6)
ssig_nodrop,nN_nodrop = gen_dataset(pt_files[:10],sample_sec=6,drop_to=0)
```

```python
basel = 50
for i,N in enumerate(nN[basel:basel+5]):
    plt.plot(ssig[basel+i])
    plt.text(50,0.5,str(N))
    plt.show()

Xtotal = np.reshape(ssig,(ssig.shape[0],ssig.shape[1],1))
Ytotal = np.array(nN)

Nrand = int(Xtotal.shape[0]/2)
print(Nrand)
#random training
#train_args = np.random.randint(Xtotal.shape[0], size=Nrand)
#targeted training
train_args = []
targets = range(1,15)
for t in targets:
    t_fil = (Ytotal-t == 0)
    tval = np.ones(Ytotal.shape)
    tval[~t_fil] = 0
    sel_args = np.flatnonzero(tval)
    end_bound = min(len(sel_args),Nrand/len(targets))
    for i,a in enumerate(sel_args):
        if i > end_bound:
            break
        train_args.append(a)
    
X_train_cnn = Xtotal[train_args]
Y_train_cnn = Ytotal[train_args]

X_valid_cnn = Xtotal
Y_valid_cnn = Ytotal

X_full = np.reshape(ssig_nodrop,(ssig_nodrop.shape[0],ssig_nodrop.shape[1],1))
Y_full = np.array(nN_nodrop)
```

```python
model = Sequential()
#model.add(MaxPooling1D(pool_size=4, input_shape = (N,1))) #sigmoid
model.add(Conv1D(filters = 128, kernel_size = 32, activation = 'sigmoid', padding = "causal"))#, input_shape = (N,1)))
model.add(MaxPooling1D(pool_size=32))
model.add(Dropout(rate = 0.25))
model.add(Flatten())

#model.add(Dense(128, activation = 'sigmoid'))

model.add(Dense(64, activation = 'sigmoid'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))#, activation = 'sigmoid'))

model.compile(
            loss = 'mean_absolute_percentage_error',#'mean_squared_error',#
            optimizer = 'nadam',
            metrics = ['mse'])
```

```python
N = X_train_cnn[0].shape[0]
min_RR_samp = int(min_RR*fs)
print("Input size = %d\nPool layer size = %d"%(N,min_RR_samp))

model = Sequential()
model.add(Conv1D(filters = 128, kernel_size = 32, activation = 'relu', padding = "causal", input_shape = (N,1)))
model.add(MaxPooling1D(pool_size=min_RR_samp,padding="same"))
model.add(Conv1D(filters = 64, kernel_size = 8, activation = 'relu', padding = "causal"))
model.add(MaxPooling1D(pool_size=int(0.5*min_RR_samp),padding="same"))
model.add(Dropout(rate = 0.1))
model.add(Flatten())

#model.add(Dense(128, activation = 'sigmoid'))

model.add(Dense(64, activation = 'sigmoid'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))#, activation = 'sigmoid'))

model.compile(
            loss = 'mean_absolute_percentage_error',#'mean_squared_error',#
            optimizer = 'nadam',
            metrics = ['mse'])
```

```python
model.fit(X_train_cnn, Y_train_cnn, batch_size = 10, epochs= 100, verbose = 1)
model.save('keras_model.h5', include_optimizer=False)
```

```python
y_train_preds_cnn = model.predict(X_train_cnn,verbose = 1)
y_valid_preds_cnn = model.predict(X_valid_cnn,verbose = 1) 
y_full_preds = model.predict(X_full,verbose = 1)

#Test with a single vector of zeros to see
#what the model returns
Y_test = model.predict(np.zeros((1,2160,1)))
print("Dummy input test -- zero input yields %.3f"%(Y_test))
```
